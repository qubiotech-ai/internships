## Crear dataset CSIRO rCMBs para test

```python
import os
import glob
import time
import nibabel as nib

# ==========================================
# 1. CONFIGURACION
# ==========================================
# Ruta origen (AIBL con rCMBs reales)
INPUT_ROOT_DIR = "/media/PORT-DISK/Practicas/MicroBleeds_Generation/CSIRO_dataset/rCMB_DefiniteSubject"

# Ruta destino para test (La que estabas utilizando en el log de error)
OUTPUT_DIR = "/media/PORT-DISK/Practicas/MicroBleeds_Generation/CSIRO_Real_CMBs_Test/imagesTs"

# Crear estructura de carpetas
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==========================================
# 2. EJECUCION DEL PIPELINE MASIVO (TEST)
# ==========================================
def prepare_test_dataset():
    print("Iniciando Pipeline de Estructuracion para Test...")
    
    nifti_files = glob.glob(os.path.join(INPUT_ROOT_DIR, "**/*.nii.gz"), recursive=True)
    print(f"Archivos encontrados: {len(nifti_files)}")

    if len(nifti_files) == 0:
        print("No se encontraron archivos en el directorio origen.")
        return

    archivos_procesados = 0

    for idx, file_path in enumerate(nifti_files):
        start_time = time.time()
        
        filename = os.path.basename(file_path)
        
        # Generar nombre con formato nnU-Net
        basename = filename[:-7] if filename.endswith('.nii.gz') else filename
        out_name_img = f"{basename}_0000.nii.gz"
        
        dst_path = os.path.join(OUTPUT_DIR, out_name_img)
        
        print(f"\n[{idx+1}/{len(nifti_files)}] {filename} -> {out_name_img}")
        
        try:
            # Cargar y guardar usando nibabel para evitar el Errno 1 del disco externo
            nii = nib.load(file_path)
            nib.save(nii, dst_path)
            
            archivos_procesados += 1
            elapsed = time.time() - start_time
            print(f"  Ok ({elapsed:.2f}s)")
            
        except Exception as e:
            print(f"  ERROR procesando {filename}: {e}")

    print("-" * 50)
    print(f"TODO LISTO. {archivos_procesados} archivos preparados en: {OUTPUT_DIR}")
    print(" -> Siguiente paso: nnUNetv2_predict -i " + OUTPUT_DIR)

if __name__ == "__main__":
    prepare_test_dataset()
```
